```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
set1 = np.random.poisson(5, 1000)
```

```python
sns.distplot(set1)
```

```python
set2 = np.random.poisson(7,1000)
set3 = np.random.poisson(9, 1000)
```

```python
sns.distplot(set2)
```

```python
sns.distplot(set3)
```

```python
## combined plot for the three histograms
sns.distplot(set1)
sns.distplot(set2)
sns.distplot(set3)
```

```python
## Poisson distribution expresses the probability of a given number of events occurring in a fixed 
## interval of time or space if these events occur with a known constant rate and independently of the time since the last event.
## Here lambda=5,7,9 for the graphs in order. We observe with increasing 'lambda' value the peeks shift more towards the right
```

```python
mu, sigma = 0, 0.1
set1_n = np.random.normal(mu, sigma, 1000)
sns.distplot(set1_n)
```

```python
mu, sigma = 0, 0.23
set2_n = np.random.normal(mu, sigma, 1000)
sns.distplot(set2_n)
```

```python
mu, sigma = 0, 0.45
set3_n = np.random.normal(mu, sigma, 1000)
sns.distplot(set3_n)
```

```python
## The combined graph of all three sets of numbers.
sns.distplot(set1_n)
sns.distplot(set2_n)
sns.distplot(set3_n)
```
